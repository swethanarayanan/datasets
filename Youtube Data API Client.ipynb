```python
def youtube_search_id(q, dk=DEVELOPER_KEY, max_results=1,order="relevance", token=None, location=None, location_radius=None):

    youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION,developerKey=dk)

    search_response = youtube.search().list(
    q=q,
    type="video",
    pageToken=token,
    order = order,
    part="id", # Part signifies the different types of data you want 
    maxResults=max_results,
    location=location).execute()
    
    videoId = []
    
    for search_result in search_response.get("items", []):
        if search_result["id"]["kind"] == "youtube#video":
            videoId.append(search_result['id']['videoId'])
        
    youtube_dict = {'videoId':videoId}

    return youtube_dict
```

### Get stats given youtube link

```python
dk_id = 40 
```

```python
def get_authenticated_service(dk=DEVELOPER_KEY):
    return build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION,developerKey=dk)
```

```python
def videos_list_by_id(dk_id,**kwargs):
    try:
        return get_authenticated_service(developer_key_list[dk_id]).videos().list(**kwargs).execute()
    except HttpError:
          if dk_id + 1 < dk_size:
            return videos_list_by_id(dk_id + 1,**kwargs)
          else:
            return
```

```python
dfs.shape
```

```python
n = 50  #chunk row size
list_dfs = [dfs[i:i+n] for i in range(starting_index,dfs.shape[0],n)]
```

```python
list_dfs
```

```python
for df in list_dfs:
    print(df.shape)
    youtubelinks = df['Youtubelink'].values.tolist()
    youtubelinks_str = ','.join(str(e) for e in youtubelinks)
    print(youtubelinks_str)
    response = videos_list_by_id(dk_id,
        part='statistics',
        id=youtubelinks_str)
    try:
        if('items' in response):
            for i in response['items']:
                if('viewCount' in i['statistics']):
                    dfs.loc[dfs['Youtubelink'] == i['id'], 'viewCount'] = i['statistics']['viewCount']

                if('likeCount' in i['statistics']):
                    dfs.loc[dfs['Youtubelink'] == i['id'], 'likeCount'] = i['statistics']['likeCount']

                #dislike may be disabled for some videos
                if('dislikeCount' in i['statistics']):
                    dfs.loc[dfs['Youtubelink'] == i['id'], 'dislikeCount'] = i['statistics']['dislikeCount']

                #comments may be disabled for some videos
                if('commentCount' in i['statistics']):
                    dfs.loc[dfs['Youtubelink'] == i['id'], 'commentCount'] = i['statistics']['commentCount']
    except:
        print(response)
```
